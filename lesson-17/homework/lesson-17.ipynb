```python
#Create a DataFrame named sales_and_expenses with columns 'Month', 'Sales', and 'Expenses', representing monthly sales and expenses data. Use below table.
dataa = {'Month':['Jan', 'Feb','Mar', 'Apr'],'Sales':[5000, 6000, 7500, 8000],'Expences': [3000, 3500,4000, 4500]}
df1 = pd.DataFrame(dataa)
df1
```

```python
#Calculate and display the maximum sales and expenses.
print(df1[["Sales", "Expences"]].max()) 
#Calculate and display the minimum sales and expenses.
print(df1[["Sales", "Expences"]].min())
print(df1[["Sales", "Expences"]].mean())
```

```python
data1 = [['Rent'	,1200,	1300,	1400,	1500],
['Utilities', 200,	220,	240,	250],
['Groceries',	300,	320,	330,	350],
['Entertainment',	150,	160,	170,	180]
]
df2 = pd.DataFrame(data1, columns=["Category", "January",	"February",	"March",	"April"]
                   )
df2

```

```python
# Calculate and display the maximum expense for each category.
df2['max_expence'] = df2[["January",	"February",	"March",	"April"]].max(axis=1)
df2
```

```python
#Calculate and display the minimum expense for each category.
df2['min_expence'] = df2[["January",	"February",	"March",	"April"]].min(axis=1)
df2
```

```python
df2.set_index("Category")
df2['mean_expence'] = df2[["January",	"February",	"March",	"April"]].mean(axis=1)
df2
```
